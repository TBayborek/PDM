```python
# Import necessary libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.transforms as transforms

from matplotlib.ticker import LinearLocator, FormatStrFormatter
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

from scipy import interpolate
```

```python
# Import data set
df_raw = pd.read_excel("Flux_resultats_python.xlsx")
print(df_raw.head(3))
print(df_raw.tail(3))
```

```python
print(df_raw.head())

d_bp = df_raw.iloc[1:,0]
d_bp.index = range(0,len(d_bp))

ei_1i2 = df_raw.iloc[0,1:6]
ei_ti= df_raw.iloc[0,8:]

df_1i2 = df_raw.iloc[1:,1:6]
df_ti = df_raw.iloc[1:,8:]

df_1i2.index = d_bp
df_1i2.columns = ei_1i2

df_ti.index = d_bp
df_ti.columns = ei_ti

print("\n1 isthme 2:\n",df_1i2)
print("\ntout isthme:\n",df_ti)
```

```python
to_plot = ["Tous isthmes","1 isthme sur 2"]
ei = [ei_ti,ei_1i2]
df_i = [df_ti,df_1i2]

for i, elem_to_plot in enumerate(to_plot):

    rrange0, rrange1 = np.meshgrid(ei[i],d_bp) #Create a meshgrid to 3D plot
    cmap_list = [cm.PiYG, cm.coolwarm]
    ax = Axes3D(plt.figure(figsize=(8, 6)))
    ax.plot_surface(rrange0, rrange1, df_i[i], cmap=cm.coolwarm, label="1 isthme sur 2")
    ax.set_xlabel("épaisseur isthmes [mm]")
    ax.set_ylabel("Distance bague-pôle [mm]")
    ax.set_zlabel("$\hat{\phi}_{1coil}$ [E-6 Wb]")
    ax.set_zlim3d([35,50])
    ax.view_init(elev=15, azim=315)
    plt.title(elem_to_plot)
    plt.margins(0)
    plt.savefig("Flux_3dplot_"+str(elem_to_plot)+".pdf")
    plt.show()
```

```python
Temp = [-40, 20, 80, 160]
flux_w10 = [51.57, 45.57, 40.59, 32.69]


flux_w05_i = np.array([49.65, 43.34, 38.09, 29.71])
flux_w075_i = np.array([47.95, 41.61, 36.33, 27.89])

flux_ti_temp = []
for i in  range(0,len(flux_w05_i)):
    g = interpolate.interp1d([0.5, 0.75], [flux_w05_i[i],flux_w075_i[i]], fill_value = "extrapolate")
    flux_ti_temp.append(np.round(g(0.71),2))
    
plt.plot(Temp,flux_w10,'o-', label= "un isthme sur deux")
plt.plot(Temp,flux_ti_temp,'o-', label= "tous isthmes")
plt.xlabel("Température [°C]")
plt.ylabel("$\hat{\phi}_{1coil}$ [E-6 Wb]")
plt.xlim([min(Temp),max(Temp)])
plt.ylim([0,55])
plt.xticks(Temp)

plt.legend(loc="best")
plt.grid()
plt.savefig("Flux_Temp_variation.pdf")
plt.show()

print("Ti",flux_ti_temp)
print("\n1i2",flux_w10)
print("Ecart rel",np.round((np.array(flux_ti_temp)-np.array(flux_w10))/np.array(flux_w10)*100,2))
```

```python
to_plot = [["rotor IPM évidé", "rotor bague $D_i = 25mm$", "rotor bague $D_i = 18mm$"]]
to_plot.append(["rotor IPM évidé", "rotor bague $D_i = 28mm$", "rotor bague $D_i = 21mm$"])
to_plot_D = ["D32", "D36"]

Temp = [-40, 20, 80, 160]
Npp = 5

R01coil = np.array([44.55, 50.43])*1E-6 #[D32, D36]

phi_D32 = np.array([[56.75, 50.35, 45.01, 36.4], # Phi [évidé, bague Di=25mm, bague Di=18mm]
               [46.9, 41.45, 37.1, 30.55],
               [57.05, 50.44, 45.14, 37.19]])*1E-6

phi_D36 = np.array([[64.70, 57.41, 51.32, 41.56], # Phi [évidé, bague Di=28mm, bague Di=21mm]
               [53.59, 47.37, 42.39, 34.91],
               [64.23, 56.79, 50.82, 41.86]])*1E-6

phi_D = [phi_D32,phi_D36]

Km = []
for j, phi_D_elem in enumerate(phi_D):
    for i, elem_to_plot in enumerate(to_plot[j]):
        plt.plot(Temp, phi_D_elem[i]*1E+6,"o-", label = elem_to_plot)

    plt.xlabel("Température [°C]")
    plt.ylabel("$\hat{\phi}_{1coil}$ [E-6 Wb]")

    plt.xticks(Temp)
    plt.xlim([min(Temp),max(Temp)])
    plt.ylim([30,None])
    plt.legend(loc="best")
    plt.grid()
    plt.savefig("Comparaison_IPM_Bague_%s.pdf"%(to_plot_D[j]))
    plt.show()

    Km.append(np.round(np.sqrt(3)*Npp*phi_D[j]/np.sqrt(R01coil[j]),5))
print("Km D32: \n", Km[0])
print("Km D36: \n", Km[1])

Ecart_rel = np.round((Km[1]-Km[0])/Km[0]*100,1)
print("Ecart_relatif: \n", Ecart_rel)
```

```python
### Import data set
time_const_perc = 1-1/np.e

# To Read
phase_single = ["U1","V1","W1","U2","V2","W2"]
phase_etoile = ["UVW1","VWU1","WUV1","UVW2","VWU2","WUV2"]
phase_etoile_T = ["UVWT","VWUT","WUVT"]
phase = phase_single + phase_etoile + phase_etoile_T

to_read = phase

R = pd.DataFrame(columns = to_read,index=[0])
R[phase_single] = 0.495
R[phase_etoile] = 0.74
R[phase_etoile_T] = 0.37

L = pd.DataFrame(columns = to_read, index = ["d","q"])

Var_dict_mes = {} #Dictionnary containing mesured variables
s=0
plt.figure("d")
plt.figure("q")
for i, read_elem in enumerate(to_read):
    for j, axis_elem in enumerate(["d","q"]):
        if (read_elem not in phase_single) & (s==0):
            plt.show()
            s = 1
        
        plt.figure(axis_elem)
        df_raw = pd.read_csv("Mesures_Inductances/%s_%s.csv"%(read_elem,axis_elem),skiprows=1)
        
        first_elem = (df_raw["Ampere"]>0.02).idxmax()
        number_elem = 100
        n_elem_shift = 0
        df = df_raw.iloc[first_elem:(first_elem+number_elem),:]
        time = df["second"]-df["second"].iloc[0]#np.arange(0,len(df["second"])*time_step,time_step)
        Ampere = df["Ampere"]
        plt.plot(time,Ampere-Ampere.iloc[0],label = read_elem+"_"+axis_elem)
        
        plt.xlim([0,0.03])
        #plt.ylim(bottom=0)
        plt.grid(True)
        plt.legend()
        
        tau = time[(Ampere>(max(Ampere)*time_const_perc)).idxmax()]
        L.loc[axis_elem,read_elem] = tau*R[read_elem][0]


plt.show()

print("L: ",L)
```
